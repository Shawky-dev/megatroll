```python
import numpy as np
from ultralytics import YOLO
import cv2
```

```python
model = YOLO("C:/Users/zeyad/Master/1_Projects/1_Active/MegaTroll/ai_training/bottle_ds_training/runs/detect/train-9/weights/best.pt")

cap = cv2.VideoCapture("http://192.168.43.1:8080/video")

while True:
    ret, frame = cap.read()
    
    if not ret:
        print("No results")
        break
    
    results = model.track(
        frame,
        persist=True,
        tracker="./tracker_conf.yaml",
    )
    
    annotated_frame = results[0].plot()
    cv2.imshow("Detection", annotated_frame)
    if cv2.waitKey(1) & 0xFF == ord('q'):
        break
    
cv2.destroyAllWindows()
```
